Project No-1 Machine Learning project for real estate domain :

Problem statement - The real estate market, a cornerstone of the global economy, has always been characterized by its ever-shifting landscape. For centuries, buyers and sellers alike have been in pursuit of accurately determining the value of a property.

It is your job to predict the sales ‘price’ for a new house. Identify the dependent and independent parameters and create a machine learning model which accurately predicts the price of a house by considering certain features as given in the dataset. Also check the accuracy you obtain from the machine learning model you create

```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('/content/Housing Project ML.csv')
```

```python
data.head()
```

```python
data['mainroad']=data['mainroad'].apply({'yes':1,'no':0}.get)
data['guestroom']=data['guestroom'].apply({'yes':1,'no':0}.get)
data['basement']=data['basement'].apply({'yes':1,'no':0}.get)
data['hotwaterheating']=data['hotwaterheating'].apply({'yes':1,'no':0}.get)
data['airconditioning']=data['airconditioning'].apply({'yes':1,'no':0}.get)
data['prefarea']=data['prefarea'].apply({'yes':1,'no':0}.get)
data['furnishingstatus']=data['furnishingstatus'].apply({'furnished':1,'semi-furnished':2,'unfurnished':3}.get)
```

```python
data.head()
```

```python
x=data[['area','bedrooms','bathrooms','stories','mainroad','guestroom','basement','hotwaterheating','airconditioning','parking','prefarea','furnishingstatus']]
y=data['price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.80)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_pred=model.predict(x_test)
```
